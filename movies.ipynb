```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

Imports NLP libraries like nltk, stopwords, lemmatizer, and regex.

```python
nltk.download('stopword')
nltk.download('wordnet')
```

Downloads necessary NLTK data (stopwords and wordnet).

```python
nltk.download('stopwords')
stop_word=set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

Loads stopwords and initializes lemmatizer.

```python
def process_text(text):
  #lower case
    text = str(text).lower()
    #remove puctuation
    text=re.sub('[^a-zA-Z]',' ',text)
    words=text.split()
    #remove stop word
    words = [word for word in words if word not in stop_word]
    #lemitize
    words=[lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)
```

Defines function to clean text (lowercase, remove symbols, etc.).

```python
df['tags']= df['tags'].apply(process_text)
```

Applies text preprocessing function to 'tags' column
